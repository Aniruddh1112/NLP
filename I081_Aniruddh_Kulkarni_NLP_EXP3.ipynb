# To Do:
obtain counts for each genre of interest. use NLTKs support
for conditional frequency distributions. 

```python
cfd = nltk.ConditionalFreqDist(
           (genre, word)
           for genre in brown.categories()
           for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

Ans: The most frequent modal in the news genre is 'will', while the most frequent modal in the romance genre is 'could'. Using this method we can make models to analyse word counts and hence distinguish genres

# Wordlist Corpora
NLTK includes some corpora that are nothing more than wordlists. We can use it to find unusual or misspelled words in a text corpus

Filtering a text: This program computes the vocabulary of a text, then removes all items that occur in an existing wordlist, leaving just the uncommon or misspelled words.

```python
def unusual_words(text):
  text_vocab=set(w.lower() for w in text if w.isalpha())
  english_vocab=set(w.lower() for w in nltk.corpus.words.words())
  unusual=text_vocab.difference(english_vocab)
  return sorted(unusual)
```

```python
print(nltk.corpus.words.words()[:25])
```

# Write a function in python to compute what fraction of words in a text are not in the stopwords list and check the fraction in gutenberg,reuters,brown and genesis corpus.

```python
import string
from nltk.tokenize import word_tokenize
import re
from nltk.corpus import stopwords

def check(raw):
  stop_words_nltk=stopwords.words('english')
  print('Number of stopwords:',len(stop_words_nltk))
  print()
  corpus_lower_3 = raw
  corpus_new_3 = re.sub(r'\d+', ' ', corpus_lower_3)
  corpus_new_3 = corpus_new_3.translate(str.maketrans("", "", string.punctuation))
  corpus_new_3 = ' '.join([token for token in corpus_new_3.split()])
  tokenized_corpus_nltk_3 = word_tokenize(corpus_new_3)
  print("Total Number of words in given Corpus: ",len(tokenized_corpus_nltk_3))
  print()
  tokenized_corpus_without_stopwords_3 = [i for i in tokenized_corpus_nltk_3 if not i in stop_words_nltk]
  print("Number of words in corpus which are not in stopword list:",len(tokenized_corpus_without_stopwords_3))
  print("Content fraction is:",(len(tokenized_corpus_without_stopwords_3)/len(raw))) 
```

```python
raw1=nltk.corpus.gutenberg.raw()
check(raw1)
```

```python
raw2=nltk.corpus.reuters.raw()
check(raw2)
```

```python
raw3=nltk.corpus.brown.raw()
check(raw3)
```

```python
raw4=nltk.corpus.genesis.raw()
check(raw4)
```
